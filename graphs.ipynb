```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

filename = 'econ_2025-07-05_util_2025-07-06_09-30-09.csv'
data_dir = 'data/'
df = pd.read_csv(data_dir + filename)

df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
cols = df.columns
print(cols)

cols = ['population', 'hours_used', 'demand_qty', 'Equipment', 'Inventory', 'Land In Use']
df = df[cols]
df_cap = df.iloc[:,1:].div(df['population'], axis=0)
# print(df)
# df['per_capita'] = df['Land In Use'] / df['population']
# df.drop([cols[1:]], axis=1, inplace=True)

# Plot all numeric columns over time
# plt.style.use('dark_background')
plt.style.use('bmh')
df.plot(figsize=(10, 6))
df_cap.plot(figsize=(10, 6))
plt.title('Title')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.grid(True)
plt.legend(title='Econ Sim')
# plt.legend(loc='upper center')
plt.tight_layout()
plt.show()

print(plt.style.available)
```
